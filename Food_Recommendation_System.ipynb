```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
```

```python
# Count the total number of ratings in the dataset
numerator = Data.select("Rating").count()

# Count the number of distinct userIds and distinct placeid
num_users = Data.select("UserID").distinct().count()
num_Places = Data.select("PlaceID").distinct().count()

# Set the denominator equal to the number of users multiplied by the number of recepies
denominator = num_users * num_Places

# Divide the numerator by the denominator
sparsity = (1.0 - (numerator *1.0)/denominator)*100
print("The ratings dataframe is ", "%.2f" % sparsity + "% empty.")
```

```python
# Group data by userId, count ratings
userId_ratings = Data.groupBy("UserID").count().orderBy('count', ascending=False)
userId_ratings.show()
```

```python
# Group data by userId, count ratings
placeID_ratings = Data.groupBy("PlaceID").count().orderBy('count', ascending=False)
placeID_ratings.show()
```

```python
# Import the required functions
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
```

```python
# Create test and train set
(train, test) = Data.randomSplit([0.8, 0.2], seed = 1234)

# Create ALS model
als = ALS(userCol="UserID", itemCol="PlaceID", ratingCol="Rating", nonnegative = True, implicitPrefs = False, coldStartStrategy="drop")

# Confirm that a model called "als" was created
type(als)
```

```python
# Add hyperparameters and their respective values to param_grid
param_grid = ParamGridBuilder() \
            .addGrid(als.rank, [1, 5, 10, 15]) \
            .addGrid(als.regParam, [.01, .05, .10, .15]) \
            .build()
            #             .addGrid(als.maxIter, [5, 50, 100, 200]) \

           
# Define evaluator as RMSE and print length of evaluator
evaluator = RegressionEvaluator(metricName="rmse", labelCol="Rating", predictionCol="prediction") 
print ("Num models to be tested: ", len(param_grid))
```

```python
# Build cross validation using CrossValidator
cv = CrossValidator(estimator=als, estimatorParamMaps=param_grid, evaluator=evaluator, numFolds=5)

# Confirm cv was built
print(cv)
```

```python
#Fit cross validator to the 'train' dataset
model = cv.fit(train)
```

```python
#Extract best model from the cv model above
best_model = model.bestModel
```

```python
# Print best_model
print(type(best_model))

# Complete the code below to extract the ALS model parameters
print("**Best Model**")

# # Print "Rank"
print("  Rank:", best_model._java_obj.parent().getRank())

# Print "MaxIter"
print("  MaxIter:", best_model._java_obj.parent().getMaxIter())

# Print "RegParam"
print("  RegParam:", best_model._java_obj.parent().getRegParam())
```

```python
# View the predictions
test_predictions = best_model.transform(test)
RMSE = evaluator.evaluate(test_predictions)
print(RMSE)
```
